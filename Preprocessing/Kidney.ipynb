RENAME COLUMNS

```python
df.rename(columns={'bp':'blood_pressure','sg':'specific_gravity','al':'albumin','su':'sugar','rbc':'red_blood_cells','pc':'pus_cell','pcc':'pus_cell_clumps','ba':'bacteria','bgr':'blood_glucose_random','bu':'blood_urea','sc':'serum_creatinine','sod':'sodium','pot':'potassium','hemo':'hemoglobin','pcv':'packed_cell_volume','wc':'white_blood_cell_count','rc':'red_blood_cell_count','htn':'hypertension','dm':'diabetes_mellitus','cad':'coronary_artery_disease','appet':'appetite','pe':'pedal_edema','ane':'anemia'},inplace=True)
```

LET'S FIRST UNDERSTAND UNIQUE VALUES IN EACH OF THE OBJECT COLUMNS FOR ENCODING OR CONVERTING THE DTYPES

```python
for i in df.select_dtypes(include=["object"]).columns:
  print(i,df[i].unique(),df[i].nunique())
```

LET'S REMOVE ID COLUMN AND CONVERT PCV,WC,RC TO FLOAT BEFORE PROCEEDING WITH ENCODING

```python
df.drop('id',axis=1,inplace=True)
```

```python
df['packed_cell_volume']=df['packed_cell_volume'].replace('?',np.nan).astype(float)
df['white_blood_cell_count']=df['white_blood_cell_count'].replace('?',np.nan).astype(float)
df['red_blood_cell_count']=df['red_blood_cell_count'].replace('?',np.nan).astype(float)
```

```python
df.info()
```

LET'S IMPUTE WITH MODE FOR CATEGORICAL COLUMNS AND WITH MEAN FOR NUMERICAL COLUMNS

```python
numerical_columns = df.select_dtypes(include=[np.number]).columns
categorical_columns = df.select_dtypes(include=[object]).columns
print(numerical_columns)
print(categorical_columns)
```

```python
for col in numerical_columns:
  df[col]=df[col].fillna(df[col].median())
```

```python
for col in categorical_columns:
  df[col]=df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.describe()
```

LET US ENCODE THE CATEGORICAL COLUMNS

```python
from sklearn.preprocessing import OrdinalEncoder
encoder=OrdinalEncoder()
```

```python
for col in categorical_columns:
  df[col]=encoder.fit_transform(df[[col]])
```

```python
df
```

```python
df.describe()
```

LET US ANALYZE DUPICATES AND MISSING VALUES

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

LET US SEE THE TARGET VARIABLE DISTRIBUTIONS

```python
x=df.drop('classification',axis=1)
y=df['classification']
```

```python
y.value_counts()
```

LET'S USE RESAMPLING TECHNIQUE

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
sample=SMOTE()
x_sample,y_sample=sample.fit_resample(x,y)
```

```python
y_sample.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sample,y_sample,test_size=0.3)
```

CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_split=2, min_samples_leaf=1)
cv_scores = cross_val_score(model, x_sample, y_sample, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean accuracy: {cv_scores.mean():.4f}")
print(f"Standard deviation: {cv_scores.std():.4f}")
```

TRAINING OUR MODEL

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
