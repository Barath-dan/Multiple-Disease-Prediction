STATISTICAL ANALYSIS

We will not delete/transform outliers because they can represent rare but possible cases of parkinson prediction.


LETS DO A CORRELATION ANALYSIS TO DETERMINE THE STRENGTH OF INFLUENCE OF OTHER COLUMNS TO TARGET COLUMN

```python
corr_matrix=df.corr()
corr_matrix
```

LET'S CONSIDER COLUMNS WITH CORRELATION VALUES BELOW 0.05 AS THRESHOLD TO DETERMINE THE STRENGTH OF RELATION THEREFORE COLUMNS WITH CORRELATION VALUE BELOW 0.05 WILL BE CONSIDERED AS WEAKLY CORRELATED


```python
plt.figure(figsize=(13, 10))
sns.heatmap(corr_matrix,annot=True, cmap='coolwarm',fmt='.2f',vmin=-1, vmax=1)
```

FROM ABOVE TABLE WE SEE THAT ALL OTHER COLUMNS HAVE MEDIUM TO STRONG CORRELATION VALUES WITH OUR TARGET COLUMN STATUS

LETS CONTINUE WITH DATA SPLITTING INTO TRAIN AND TEST DATA

```python
y=df['status']
x=df.drop('status',axis=1)
```

LET'S UNDERSTAND THE TARGET COLUMN DISTRIBUTIONS

```python
y.value_counts()
```

RESAMPLE THE DATA FOR BETTER MODEL PERFORMANCE

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
sample=SMOTE()
x_sample,y_sample=sample.fit_resample(x,y)
```

```python
y_sample.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sample,y_sample,test_size=0.2)
```

LET'S SEE AND COMPARE THE ACCURACY RESULTS OF DIFFERENT ALGORITHMS FOR OUR MODEL

```python
from sklearn.ensemble import RandomForestClassifier
rf_clf=RandomForestClassifier(max_depth=10,n_estimators=100,min_samples_split=5,min_samples_leaf=4)
rf_clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
y_pred=rf_clf.predict(x_test)
print(classification_report(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=500)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
import xgboost as xgb
```

```python
xgb_clf = xgb.XGBClassifier(n_estimators=100, max_depth=2, learning_rate=0.1, random_state=42,)
xgb_clf.fit(x_train, y_train)
```

```python
y_pred =xgb_clf.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf', C=1.0, gamma='scale', random_state=42)
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

LET US CHOOSE XGBOOST ALGORITHM AS IT PRODUCES BETTER ACCURACY COMPARED TO OTHER MODELS

CROSS VALIDATE OUR MODEL

```python
from sklearn.model_selection import cross_val_score
model=xgb_clf
cv_scores = cross_val_score(model, x_sample, y_sample, cv=5, scoring='accuracy')
print(f"Cross-validation scores: {cv_scores}")
print(f"Mean accuracy: {cv_scores.mean():.4f}")
print(f"Standard deviation: {cv_scores.std():.4f}")
```
